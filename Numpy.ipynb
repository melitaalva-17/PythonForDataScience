```python
# how can u count the frequency of given positive value in numpy array?
import numpy as np
def my_counter(array,num):
    counter=0
    for x in array:
        if num == x:
            counter=counter+1
    return counter
number=np.array([2,3,5,7,7,6])
ans =my_counter(number,7)
print(ans)
```

```python
'i', 'love', 'NumPy', 'AND', 'Akanksha mam'
```

```python

Str="I Love Numpy and Akansha Mam"
print("Upper case string is: ",Str.upper())
print("Lower case string is: ",Str.lower())
print("Capitalize the first: ",Str.capitalize())
print("Title case string is: ",Str.title())
print("Swap the case: ",Str.swapcase())
```

```python
arr = np.array(['we', 'love', 'NumPy', 'AND', 'akansha mam'], dtype=str)

upper_case_arr = np.char.upper(arr)
lower_case_arr = np.char.lower(arr)
capitalize_case_arr = np.char.capitalize(arr)
titlecase_arr = np.char.title(arr)
swapcase_arr = np.char.swapcase(arr)

print("Upper Conversion: ", upper_case_arr)
print("Lower Conversion: ", lower_case_arr)
print("Capitalize First Letter Conversion: ", capitalize_case_arr)
print("Titlecase Conversion: ", titlecase_arr)
print("Swapcase Conversion: ", swapcase_arr)
```
